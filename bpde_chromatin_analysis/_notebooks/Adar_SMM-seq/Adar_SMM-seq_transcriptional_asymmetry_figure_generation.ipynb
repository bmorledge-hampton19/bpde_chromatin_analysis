### Parse the gene bins file to a pandas data frame

```python
import os
from bpde_chromatin_analysis.helper_scripts.BPDE_DataDir import getDataDir
from chromatinfeaturesanalysis.BinInGenes import parseGeneBinData, plotGeneBins

smmSeqDataDirectory = os.path.join(getDataDir(), "Adar_smm-seq")
smmSeqGeneBinDataFilePath = os.path.join(smmSeqDataDirectory, "Adar_smm-seq_trinuc_context_mutations_gene_bins_TS_vs_NTS.tsv")
smmSeqGeneBinData = parseGeneBinData(smmSeqGeneBinDataFilePath)
smmSeqFineGrainGeneBinDataFilePath = os.path.join(smmSeqDataDirectory, "Adar_smm-seq_trinuc_context_mutations_gene_bins_TS_vs_NTS_fine-grain.tsv")
smmSeqFineGrainGeneBinData = parseGeneBinData(smmSeqFineGrainGeneBinDataFilePath)
```

### Graph the TS vs. NTS

```python
print(
    plotGeneBins(smmSeqGeneBinData, "Adar SMM-seq Transcriptional Asymmetry", yAxisLabel = "Mutations (pyrimidine)",
                 yData1 = "Coding_Strand_Counts", yData2 = "Noncoding_Strand_Counts", plotYData3Only = False,
                 flankingBinSize = 10886, flankingBinNum = 3)
)
print(
    plotGeneBins(smmSeqFineGrainGeneBinData, "Adar SMM-seq Transcriptional Asymmetry", yAxisLabel = "Mutations (pyrimidine)",
                 yData1 = "Coding_Strand_Counts", yData2 = "Noncoding_Strand_Counts", plotYData3Only = False,
                 flankingBinSize = 108, flankingBinNum = 300, geneFractionNum = 600)
)
print(
    plotGeneBins(smmSeqFineGrainGeneBinData, "Adar SMM-seq Transcriptional Asymmetry", yAxisLabel = "Mutations (pyrimidine)",
                 yData1 = "Coding_Strand_Counts", yData2 = "Noncoding_Strand_Counts", plotYData3Only = False,
                 flankingBinSize = 108, flankingBinNum = 300, geneFractionNum = 600, xlim = (-10,10))
)
```

### Graph the log ratio of TS vs. NTS

```python
print(
    plotGeneBins(smmSeqGeneBinData, "Adar SMM-seq Transcriptional Asymmetry", yAxisLabel = "Mutation Log Ratio (Pyr; TS/NTS)", ylim = (-1,1),
                 flankingBinSize = 10886, flankingBinNum = 3)
)
print(
    plotGeneBins(smmSeqFineGrainGeneBinData, "Adar SMM-seq Transcriptional Asymmetry", yAxisLabel = "Mutation Log Ratio (Pyr; TS/NTS)", ylim = (-1,1),
                 flankingBinSize = 108, flankingBinNum = 300, geneFractionNum = 600)
)
print(
    plotGeneBins(smmSeqFineGrainGeneBinData, "Adar SMM-seq Transcriptional Asymmetry", yAxisLabel = "Mutation Log Ratio (Pyr; TS/NTS)", ylim = (-1,1),
                 flankingBinSize = 108, flankingBinNum = 300, geneFractionNum = 600, xlim = (-10,10))
)
```
